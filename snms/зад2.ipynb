```python
#crude monte carlo method for task 2a)
#in this task we have a four-dimensional unit cube,in which case the integral can be reduced 
#to the product of 4 one-dimensional integrals
#
#function I = f_x * f_x1 * f_x2 * f_x3

import numpy as np #imported numpy library with alias name np 
N = 100000 #the number of function evaluations
a = 0 #the lower bound of the integral
b = 1 #the upper bound of the integral
sum = 0
#with numpy.random.uniform() method,can get random samples from uniform distribution 
x = np.random.uniform(a,b,N)
x1 = np.random.uniform(a,b,N)
x2 = np.random.uniform(a,b,N)
x3 = np.random.uniform(a,b,N)

f_x = np.exp(x) #first integral e^x
f_x1 = np.sin(x1) #second integral sinx1
f_x2 = np.cos(x2) #third integral cosx2
f_x3 = np.log(1+x3) #forth integral log(1+x3)
sum = f_x * f_x1 * f_x2 * f_x3 #sum of multiplication 

#numpy.mean() function used to compute the arithmetic mean of the given elements
print("first integral ~ ",np.mean(f_x)*(b-a)) #approximate solution of a first 1D-integral
print("second integral ~ ",np.mean(f_x1)*(b-a)) #approximate solution of a second 1D-integral
print("third integral ~ ",np.mean(f_x2)*(b-a)) #approximate solution of a third 1D-integral
print("forth integral ~ ",np.mean(f_x3)*(b-a)) #approximate solution of a forth 1D-integral
print("approximate solution of a 4D integral with MCM ~ ",np.mean(sum)*(b-a))
```

```python
def mean_by_iteration(x):
    sum = 0.0
    results = []
    for i,v in enumerate(x):
        sum += v
        results.append(sum/(i+1))
    
    return results
```

```python
import numpy as np #imported numpy library with alias name np 
from matplotlib import pyplot as plt
```

```python
N = 10000 #the number of function evaluations
a = 0 #the lower bound of the integral
b = 1 #the upper bound of the integral
sum = 0
```

```python
#with numpy.random.uniform() method,can get random samples from uniform distribution 
x = np.random.uniform(a,b,N)

_ = plt.hist(x,bins=500)
```

```python
f_x = np.exp(x) #first integral e^x
_ = plt.hist(f_x,bins=500)
np.mean(f_x)*(b-a)
```

```python
f_x1 = np.sin(x) #second integral sinx1

plt.plot(mean_by_iteration(f_x1))
np.mean(f_x1)*(b-a)
```

```python
f_x2 = np.cos(x) #third integral cosx2
plt.plot(mean_by_iteration(f_x2))
np.mean(f_x2)*(b-a)
```

```python
f_x3 = np.log(1+x) #forth integral log(1+x3)
plt.plot(mean_by_iteration(f_x3))
np.mean(f_x3)*(b-a)
```
